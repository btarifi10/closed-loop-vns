```python
import pandas as pd

# Load the data
df = pd.read_csv('./data/vns_dataset_threshold_sim.csv')

AB_df = df[df['C_activation_level'] == 0.0]
C_df = df[df['AB_activation_level'] == 1.0]
```

```python
import matplotlib.pyplot as plt

features = ['pulse_width', 'amplitude', 'frequency']

for feature in features:
    plt.figure()
    samples = AB_df['sample_id'].unique()
    for sample in samples:
        sample_df = AB_df[AB_df['sample_id'] == sample]
        plt.scatter(sample_df[feature], sample_df['AB_activation_level'], label=f'Sample {sample}', s=5)
    plt.xlabel(feature)
    plt.ylabel('Activation')
    plt.title(f'AB Activation vs {feature}')
    plt.legend()
    plt.show()
```

```python
for feature in features:
    plt.figure()
    samples = C_df['sample_id'].unique()
    for sample in samples:
        sample_df = C_df[C_df['sample_id'] == sample]
        plt.scatter(sample_df[feature], sample_df['C_activation_level'], label=f'Sample {sample}', s=5)
    plt.xlabel(feature)
    plt.ylabel('Activation')
    plt.title(f'C Activation vs {feature}')
    plt.legend()
    plt.show()
```
